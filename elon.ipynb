# IMPORT LIBRARIES

```python

import os
import warnings
warnings.simplefilter('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread, imshow
from skimage.transform import resize
from skimage.color import rgb2gray
```

```python
elon=os.listdir("C:/elon")
```

```python
sundar=os.listdir("C:/sundar")
```

```python
bill=os.listdir("C:/bill")
```

```python
limit = 10

elon_images =[None]*limit
j=0
for i in elon:
    if(j<limit):
        elon_images[j] = imread("C:/elon/"+i)
        j+=1
    else:
        break
```

```python
limit = 10

sundar_images =[None]*limit
j=0
for i in sundar:
    if(j<limit):
        sundar_images[j] = imread("C:/sundar/"+i)
        j+=1
    else:
        break
```

```python
limit = 10

bill_images =[None]*limit
j=0
for i in bill:
    if(j<limit):
        bill_images[j] = imread("C:/bill/"+i)
        j+=1
    else:
        break
```

```python
imshow(elon_images[0])
```

```python
imshow(sundar_images[0])
```

```python
imshow(bill_images[0])
```

```python
elon_gray =[None]*limit
j=0

for i in elon:
    if(j<limit):
        elon_gray[j]=rgb2gray(elon_images[j])
        j+=1
    else:
        break
```

```python
sundar_gray =[None]*limit
j=0

for i in sundar:
    if(j<limit):
        sundar_gray[j]=rgb2gray(sundar_images[j])
        j+=1
    else:
        break
```

```python
bill_gray =[None]*limit
j=0

for i in bill:
    if(j<limit):
        bill_gray[j]=rgb2gray(bill_images[j])
        j+=1
    else:
        break
```

```python
for j in range (10):
    ee = elon_gray[j]
    elon_gray[j] = resize(ee,(512,512))
```

```python
for j in range (10):
    ss = sundar_gray[j]
    sundar_gray[j] = resize(ss,(512,512))
```

```python
for j in range (10):
    bb = bill_gray[j]
    bill_gray[j] = resize(bb,(512,512))
```

# ELON

```python
len_of_images_elon=len(elon_gray)
```

```python
image_size_elon=elon_gray[1].shape
```

```python
flatten_size_elon=image_size_elon[0]*image_size_elon[1]
```

```python
elon_gray=np.dstack(elon_gray)
```

```python
elon_gray=np.rollaxis(elon_gray,axis=2,start=0)
```

```python
elon_gray=elon_gray.reshape(len_of_images_elon,flatten_size_elon)
```

```python
elon_data=pd.DataFrame(elon_gray)
```

```python
elon_data
```

```python
elon_data["label"]="elon"
```

```python
elon_data
```

# SUNDAR

```python
len_of_images_sundar=len(sundar_gray)
```

```python
image_size_sundar=sundar_gray[1].shape
```

```python
flatten_size_sundar=image_size_sundar[0]*image_size_sundar[1]
```

```python
sundar_gray=np.dstack(sundar_gray)
```

```python
sundar_gray=np.rollaxis(sundar_gray,axis=2,start=0)
```

```python
sundar_gray=sundar_gray.reshape(len_of_images_sundar,flatten_size_sundar)
```

```python
sundar_data=pd.DataFrame(sundar_gray)
```

```python
sundar_data
```

```python
sundar_data["label"]="sundar"
```

```python
sundar_data
```

```python
great=pd.concat([elon_data,sundar_data])
```

```python
great
```

```python
from sklearn.utils import shuffle
```

```python
great_indexed = shuffle(great).reset_index()
```

```python
great_indexed
```

```python
great_in=great_indexed.drop(['index'],axis=1)
```

```python
great_in
```

```python
x = great_in.values[:,:-1]
```

```python
y = great_in.values[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
from matplotlib import pyplot as plt
```

```python
from sklearn import metrics
```

```python
accuracy = metrics.accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```
